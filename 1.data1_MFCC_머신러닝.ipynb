```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Aivle_빅프/donateacry_corpus/exp1_testing.csv')
print(data.shape)
data.tail()
```

# 실험 1 : 원본 데이터 사용
- 데이터 불균형 유지되어있음
- 대부분의 machine leaning 모델에서 hungry(최빈 클래스)만 맞춤

```python
from sklearn.model_selection import train_test_split


y = data.iloc[:,-1]
X = data.drop(columns = 'label')
X = X.astype(np.float32)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)


print(X_train.shape, y_train.shape)
print(X_test.shape, y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, precision_score, recall_score

def get_scores(classifier, X_train, X_test, y_train, y_test, **kwargs):
    model = classifier(**kwargs)
    model.fit(X_train, y_train)
    y_predict = model.predict(X_test)
    conf_matrix = confusion_matrix(y_test, y_predict)
    return model.score(X_test, y_test), \
           precision_score(y_test, y_predict, average='weighted'), \
           recall_score(y_test, y_predict, average='weighted'), \
           conf_matrix

print("    Model, Accuracy, Precision, Recall, Confusion Matrix")
accuracy, precision, recall, conf_matrix = get_scores(RandomForestClassifier, X_train, X_test, y_train, y_test, n_estimators=25, max_features=5)
print("    Random Forest:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

accuracy, precision, recall, conf_matrix = get_scores(LogisticRegression, X_train, X_test, y_train, y_test)
print("    Logistic Regression:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

accuracy, precision, recall, conf_matrix = get_scores(DecisionTreeClassifier, X_train, X_test, y_train, y_test)
print("    Decision Tree:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

accuracy, precision, recall, conf_matrix = get_scores(SVC, X_train, X_test, y_train, y_test)
print("    SVM:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

# accuracy, precision, recall, conf_matrix = get_scores(MultinomialNB, X_train, X_test, y_train, y_test)
# print("    Naive Bayes:", accuracy, precision, recall)
# print("    Confusion Matrix:\n", conf_matrix)
```

# 실험 2 : hungry 데이터 비율을 줄이기
- 데이터 불균형은 해결됨
- 데이터 정보 부족으로 제대로된 분류가 되지 않음

```python
data['label'].value_counts()
```

```python
hungry_data = data[data['label'] == 'hungry']
hungry_data = hungry_data.head(27)

filterd_data = pd.concat([hungry_data, data[data['label'] != 'hungry']])
print(filterd_data.shape)
filterd_data.head()
```

```python
filterd_data['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split


y = filterd_data.iloc[:,-1]
X = filterd_data.drop(columns = 'label')
X = X.astype(np.float32)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)


print(X_train.shape, y_train.shape)
print(X_test.shape, y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, precision_score, recall_score

def get_scores(classifier, X_train, X_test, y_train, y_test, **kwargs):
    model = classifier(**kwargs)
    model.fit(X_train, y_train)
    y_predict = model.predict(X_test)
    conf_matrix = confusion_matrix(y_test, y_predict)
    return model.score(X_test, y_test), \
           precision_score(y_test, y_predict, average='weighted'), \
           recall_score(y_test, y_predict, average='weighted'), \
           conf_matrix

print("    Model, Accuracy, Precision, Recall, Confusion Matrix")
accuracy, precision, recall, conf_matrix = get_scores(RandomForestClassifier, X_train, X_test, y_train, y_test, n_estimators=25, max_features=5)
print("    Random Forest:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

accuracy, precision, recall, conf_matrix = get_scores(LogisticRegression, X_train, X_test, y_train, y_test)
print("    Logistic Regression:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

accuracy, precision, recall, conf_matrix = get_scores(DecisionTreeClassifier, X_train, X_test, y_train, y_test)
print("    Decision Tree:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

accuracy, precision, recall, conf_matrix = get_scores(SVC, X_train, X_test, y_train, y_test)
print("    SVM:", accuracy, precision, recall)
print("    Confusion Matrix:\n", conf_matrix)

# accuracy, precision, recall, conf_matrix = get_scores(MultinomialNB, X_train, X_test, y_train, y_test)
# print("    Naive Bayes:", accuracy, precision, recall)
# print("    Confusion Matrix:\n", conf_matrix)
```

```python
import cPickle

def pickle_model(model, modelname):
    with open('models/' + str(modelname) + '.pkl', 'wb') as f:
        return cPickle.dump(model, f)

model = RandomForestClassifier()
model.fit(X,y)
pickle_model(model, "myRandomForest")
```
